```python
import duckdb
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
con = duckdb.connect("../warehouse/dev.duckdb")
```

```python
con.execute("SELECT COUNT(*) FROM fct_orders").df()
```

```python
arr_mrr_df = con.execute("""
    SELECT DATE_TRUNC('month', order_date) AS month,
           COUNT(order_id) AS order_count
    FROM fct_orders
    GROUP BY month
    ORDER BY month
""").df()
```

```python
sns.set(style="whitegrid")
sns.lineplot(data=arr_mrr_df, x="month", y="order_count")
plt.title("Monthly Order Volume")
plt.show()
```

### Insights
- Retention drops sharply after month 3 — suggests onboarding improvements needed.
- ARR growth is steady, but churn spikes in Q2.
